```python
from game_data import data
from art import logo, vs
from random import randint
from replit import clear
```

```python

def PrintComp(paranumA,userscore,tocontinue):
  while tocontinue:
    numB = randint(0,50)
    while paranumA == numB :
      numB = randint(0,50)
    print("Compare A: " + data[paranumA]["name"] + ", a " + data[paranumA]["description"] + ", from " + data[paranumA]["country"] + ".") 
    print(vs) 
    print("Against B: " + data[numB]["name"] + ", a " + data[numB]["description"] + ", from " + data[numB]["country"] + ".")
    response = input("Who has more followers? Type 'A' or 'B': ").capitalize()
    follower_countA = data[paranumA]["follower_count"]
    follower_countB = data[numB]["follower_count"]
    tocontinue = False
    if response == "A" and (follower_countA > follower_countB):
      userscore += 1
      tocontinue = True
      paranumA = numB
      clear()
      print(logo)
      print(f"You're right! Current score: {userscore}.")
    elif response == "B" and (follower_countA < follower_countB):
      userscore += 1
      tocontinue = True
      paranumA = numB
      clear()
      print(logo)
      print(f"You're right! Current score: {userscore}.")
    else:
      clear()
      print(logo)
      print(f"Sorry, that's wrong. Final score: {userscore}")
  
```

```python
if __name__ == "__main__":
  print(logo) 
  # Hold user score
  userscore = 0 
  # Choose at random 1st celebrity 
  numA = randint(0,50)
  # Choose at random 2nd celebrity
  PrintComp(numA,userscore,True)
```
